```julia
using Plots
using StatsBase
using LinearAlgebra
using Statistics
using JLD2
using Dates
using LsqFit
using StaticArrays

# data structure module (needs to be module to avoid overwriting local variables in REPL scope)
include("SpinSimParams.jl")
using .SpinSimParams

# function libraries
include("spin_sims.jl");

# for analysis of M(t)
include("MTools.jl")
using .MTools
```

```julia
# interaction
α =0:0.005:0.20
ω = 0;

# spin ensemble
nx0 = 10
ny0 = 10
nz0 = 1
nfreq = nx0*ny0*nz0;

# I gave γ, τ, dt, ψ_0, bw, and ν0 default values since we almost never change them at this point

# lattice parameters
# 2D lattice:
hlk = [1; 1];
θ = [π/2];
# 3D lattice:
#hlk = [1; 1; 1];
#θ = [π/2; π/2; π/2];

# wavevectors
#ky = collect(exp10.(LinRange(-1, 1, 10)))
#pushfirst!(ky,0)
ky = [0.0];

k = ();
for i = 1:size(ky,1)
    k = (k..., [0; ky[i]]);
end
k = [0.0, 0.0];

# make the parameter file
params = make_params(α, ω, nfreq);

# load lattice parameters
params["hlk"] = hlk;
params["θ"] = θ;
params["k"] = k;

# create lattice
@time r, idx_r = make_lattice(params);
params["r"] = r#make_tiles_R(r,params);

#
# make stencil
#

params["local_M_on"] = false;

nx = nx0#nx0^2
ny = ny0#ny0^2
nz = nz0#nz0^2

#ξ_list = [0.1, 0.5, 1, 5, 10, 9999] # correlation length, in units of lattice parameter
ξ_list = [1e10]#[0.1, 0.5, 1, 5, 10, 1e10]
M_stencil_list = []
for ξ in ξ_list
    M_stencil = zeros(nx,ny,nz)

    for a = 1:nx
        for b = 1:ny
            for c = 1:nz
                a_loc = mod(a-1+nx/2,nx) - nx/2
                b_loc = mod(b-1+ny/2,ny) - ny/2
                c_loc = mod(c-1+nz/2,nz) - nz/2

                r_loc = [a_loc, b_loc, c_loc]
                M_stencil[a,b,c] = 1/(norm(r_loc)/ξ+1)

            end
        end
    end

    #M_stencil = M_stencil./sum(M_stencil)  # the vector P is used to normalize instead
    #M_stencil[1,1,1] = 0 # no onsite correlation
    push!(M_stencil_list,M_stencil)
end

params["M_stencil"] = M_stencil_list
 
#
# jump operators (dissipation)
#

gam_0_list = exp10.(-6:0.1:-2)

Ljs = []

for gam_idx in 1:length(gam_0_list)
    gam_0 = gam_0_list[gam_idx]
    gam_1 = gam_0
    gam_2 = gam_0
    gam_3 = 0.0

    Iz = @SMatrix [1/2 0; 0 -1/2];

    nJ = 3
    Lj_list = []
    push!(Lj_list, sqrt(gam_1)* @SMatrix [0 1; 0 0]) # down to up flip
    push!(Lj_list, sqrt(gam_2)* @SMatrix [0 0; 1 0]) # up to down flip
    push!(Lj_list, sqrt(gam_3)*Iz) # phase decoherence
    push!(Ljs,Lj_list)

end
params["Lj"] = Ljs
params["gamma_0"] = gam_0_list


# set the variables to loop over
vars = ["α","Lj"];
params["vars"] = vars;
```

```julia
I, d = make_idx(vars, params)
M_list = Array{Any}(undef, d)

for i in I

    # generate temporary parameters
    tparams = make_temp_params(params, vars, i)
    #make_tiles(tparams)

    # simulate
    M_list[i] = spin_echo_sim_liouville(tparams)
    if (i[1] == 1)
        println(string(i[2],"/",length(gam_0_list)))
    end

end

params["M_list"] = M_list;
```

# Load and plot alpha vs Gamma_0 data

```julia
using JSON
using LaTeXStrings

#fname = "spinecho_data.json"
fname = "spinecho_data_alpha_and_gamma.json"

dat = Dict()
open(fname, "r") do f
    global dat
    dat=JSON.parse(f)  # parse and transform data
end

#print(keys(params_loaded))
M_list = dat["M_list"];
```

```julia
# this function has been added to MTools.jl as "getMStats". Skip to next block

I, d = make_idx(dat["vars"], dat)
gr(dpi=100)
M_stats = Dict();

α = dat["α"]
gamma_0 = log10.(dat["gamma_0"])

idx_w = 200 # number of time indices on each side to keep
sym_w = 50 # number of time indices on each side to evaluate symmetry
min_cutoff = 0.02 # value of M to assess as a "dip"
mean_range = 50 # number of time indices on each side for evaluating the mean (no longer used)

for i in I
    M = [ x["re"] + im*x["im"] for x in M_list[i[2]][i[1]] ];
    nτ = convert(Int64, round(dat["τ"]*dat["γ"]/dat["dt"]));
    t = LinRange(0, 300, size(M, 1));


    idx1 = max(2*nτ-idx_w,1)
    idx2 = min(2*nτ+idx_w,length(M))
    M_tar = broadcast(abs,M[idx1:idx2])
    t_tar = t[idx1:idx2]

    plt3 = plot(t_tar,M_tar, legend=:none,
        title=latexstring("\\alpha = ",α[i[1]], ", \\Gamma_0 = 10^{", round(gamma_0[i[2]],digits=2),"}"))

    on_sweep = true
    curr_min = 0;
    min_idx = 1
    min_idx_list = []

    for idx = 1:length(M_tar)
        #dh = M_diff[idx]
        vh = M_tar[idx]
        if on_sweep # in a region of a dip
            if vh < curr_min
                curr_min = vh
                min_idx = idx
            elseif vh > min_cutoff
                push!(min_idx_list,min_idx)
                curr_min = min_cutoff
                on_sweep = false
            end
        else # looking for next dip
            if vh < min_cutoff
                curr_min = vh
                min_idx = idx
                on_sweep = true
            end
        end
    end

    push!(min_idx_list,length(M_tar))

    max_idx_list = []
    max_vs = []
    max_sigmas = []
    max_symms = []
    for i = 1:length(min_idx_list)-1

        tar_idx = min_idx_list[i]:min_idx_list[i+1]
        max_v, max_idx = findmax(M_tar[tar_idx]);
        max_idx = max_idx+tar_idx[1]-1

        if max_v < 1.5*min_cutoff
            continue
        end

        tar_idx = max(max_idx-mean_range,min_idx_list[i]):min(max_idx+mean_range,min_idx_list[i+1])

        symm_dist = min(sym_w,max_idx-1,length(M_tar)-max_idx)
        left_idx = max_idx-symm_dist:max_idx-1
        right_idx = max_idx+1:max_idx+symm_dist

        #mean_h = sum(tar_idx.*M_tar[tar_idx])/sum(M_tar[tar_idx])

        sigma_h = sqrt( sum(M_tar[tar_idx].*(t_tar[tar_idx] .- t_tar[max_idx]).^2)./sum(M_tar[tar_idx]) )
        lM = M_tar[left_idx];
        rM = reverse(M_tar[right_idx]);
        symm_h = sum(lM.*rM)/sqrt(sum(lM.*lM)*sum(rM.*rM))

        push!(max_idx_list,max_idx)    
        push!(max_sigmas,sigma_h)
        push!(max_vs,max_v)
        push!(max_symms,symm_h)

    end

    plot!([t_tar[1], t_tar[length(M_tar)]],min_cutoff .+ [0, 0],color=RGB(0,0,0),linestyle=:dot) 
    plot!([t_tar[1], t_tar[length(M_tar)]],[0, 0],color=RGB(0,0,0)) 

    for idx in min_idx_list
        plot!(t_tar[idx] .+ [0,0],[0, .5],color=RGB(0,0,0),linestyle=:dash) 
    end
    for idx in 1:length(max_idx_list)
        max_idx = max_idx_list[idx]
        sg = max_sigmas[idx]
        v = max_vs[idx]
        sym = max_symms[idx]
        v_shift = 0.015;
        plot!(t_tar[max_idx] .+ [0,0],[0, v],color=RGB(1,0,0)) 
        plot!(t_tar[max_idx] .+ [-sg,sg],[v, v],color=RGB(1,0,0)) 
        annotate!(t_tar[max_idx],v+v_shift,text(round(sym,digits=3),:center, 9, color=RGB(0,0,0)))


    end
    
    if (i[1] == 33 && i[2] == 20) # specific values of alpha (i[1]) and gamma (i[2]) of interest
        display(plt3)
        #fname = "M_analysis3_step5.png"
        #savefig(plt3,fname)
    end
    
    dict_here = Dict("min_idx"=>min_idx_list,"max_idx"=>max_idx_list,"sigmas"=>max_sigmas,"vals"=>max_vs,"symms"=>max_symms, "M"=>M_tar, "t"=>t_tar)
    M_stats[string(i)] = dict_here

end
```

```julia

M_stats = getMstats(dat)

α = dat["α"]
gamma_0 = log10.(dat["gamma_0"])

M_max_surf = zeros(length(α),length(gamma_0))
symm_surf = zeros(size(M_max_surf))
num_peaks_surf = zeros(size(M_max_surf))
ratio_surf = zeros(size(M_max_surf))

I, d = make_idx(dat["vars"], dat)
for i in I
    alpha_h = α[i[1]]
    gamma_h = gamma_0[i[2]]
    stats_h = M_stats[string(i)]
    min_idx = stats_h["min_idx"]
    max_idx = stats_h["max_idx"]
    sigmas = stats_h["sigmas"]
    vals = stats_h["vals"]
    symms = stats_h["symms"]
    M = stats_h["M"]
    t = stats_h["t"]
    
    if length(max_idx) == 0
        continue
    end
    peak_idx = findall(max_idx .== maximum(max_idx))
    
    M_max_surf[i] = maximum(vals)
    symm_surf[i] = 1 - symms[peak_idx][1]
    num_peaks_surf[i] = length(vals)
    ratio_surf[i] = vals[peak_idx][1]/sigmas[peak_idx][1]
end


p1 = heatmap(gamma_0,α,M_max_surf,title=latexstring("M_\\textrm{max}"))
p2 = heatmap(gamma_0,α,symm_surf,title=latexstring("\\textrm{Asymm}"))
p3 = heatmap(gamma_0,α,num_peaks_surf,title=latexstring("N_\\textrm{peaks}"))
p4 = heatmap(gamma_0,α,ratio_surf,title=latexstring("M_\\textrm{max}/\\sigma"))

ylabel!(p1,latexstring("\\alpha"))
ylabel!(p3,latexstring("\\alpha"))

xlabel!(p3,latexstring("\\log\\Gamma_0"))
xlabel!(p4,latexstring("\\log\\Gamma_0"))

l = @layout grid(2,2)

plt = plot(p1,p2,p3,p4, layout = l, size=(800,600))
display(plt)

fname = "alpha_gamma_sweep_stats.png";
#gr(dpi=400)
#savefig(plt,fname)
```
