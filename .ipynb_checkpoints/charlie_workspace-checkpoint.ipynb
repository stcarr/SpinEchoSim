```julia
using Plots
using StatsBase
using LinearAlgebra
using Statistics
using JLD2
using Dates
using LsqFit
using StaticArrays
using JSON

# data structure module (needs to be module to avoid overwriting local variables in REPL scope)
include("SpinSimParams.jl")
using .SpinSimParams

# function libraries
include("spin_sims.jl");

#= for analysis of M(t)
include("MTools.jl")
using .MTools
=#
```

# Set Independent Variables

```julia
# interaction
α = 0.13
ω = 0;

# number of frequencies
n_list = ();
for i = 1:20
    temp = (10*i, 10*i)
    n_list = (n_list..., temp)
end

# make the parameter file
params = make_params(α, ω, n_list[1]);
```

```julia
# make a lattice and stencil

r_list = ()
spin_idx_list = ()

for n in n_list
    
    hlk = [1; 1]./n[1]
    θ = [π/2]
    r, spin_idx = make_lattice(hlk, θ, n, false);
    
    r_list = (r_list..., r)
    spin_idx_list = (spin_idx_list..., spin_idx)
    
end

k_list = ()
for ky in exp10.(LinRange(-2, 2, 100))
    k_list = (k_list..., [0; ky])
end

#=
stencil_list = ()
for ξ_temp in ξ
    stencil = make_stencil(hlk, θ, n, r, ξ_temp, false);
    stencil_list = (stencil_list..., stencil)
end
=#
    
tile_dims = (10,10)

# params["hlk"] = hlk
# params["θ"] = θ
# params["ξ"] = ξ
params["r"] = r_list
params["k"] = k_list
params["n"] = n_list
params["pbc"] = false
params["spin_idx"] = spin_idx_list
# params["M_stencil"] = stencil_list
params["local_M_on"] = false
params["tile_dims"] = (10,10)
params["tiles"] = true;
```

# Set Options
1. "integrated_sampling": if true, use integrated sampling; if false, use discrete sampling (default = false).
2. "tiles": if true, use tile approach to generating lattice (default = false).  Must provide **params["r_tiles"] and params["tile_norm"]**.  "r_tiles" specifies the positions of the tiles, and "tile_norm" is a rescaling parameter to make sure the tiles do not overlap.
3. "local_M_on": if true, implement correlation length (default = false).  Must provide **params["M_stencil"]**.
4. $\theta$ is defined as follows: $\theta$[1] is the angle between hlk[1] and hlk[2], $\theta$[2] is the angle between hlk[1] and the projection of hlk[3] into the plan of hlk[1] and hlk[2], and $\theta$[3] is the angle between hlk[3] and the axis perpendicular to the plan created by hlk[1] and hlk[2] (ie if hlk[1] and hlk[2] are considered in the x-y plane, $\theta$[3] is the angle from the z axis to hlk[3])

```julia
# set the variables to loop over
params["integrated_sampling"] = true;
vars = [("r", "spin_idx", "n"), "k"]
params["vars"] = vars;
```

```julia
idx_list, d = make_idx(vars, params)
M_list = Array{Any}(undef, d)

for i in idx_list

    # generate temporary parameters
    tparams = make_temp_params(params, vars, i)

    # simulate
    @time M_list[i] = spin_echo_sim(tparams)

end


#= 
@save "converge_vs_k_r_M_list.jld2" M_list;

using JSON
fname = "converge_vs_k_r_params.json"
open(fname,"w") do f
    JSON.print(f, params)
end
=#
```

```julia
rmse = Array{Any}(undef, 10, 100)

for i = 1:10
    for j = 1:100
        
        ref = broadcast(abs, M_list[10, j])
        test = broadcast(abs, M_list[i, j])
        δ = sqrt(sum((test - ref).^2))
        rmse[i,j] = δ
        
    end
end

heatmap(exp10.(LinRange(-2, 2, 100)), LinRange(10, 100, 10), rmse, xscale=:log10)
# heatmap(rmse)
ylabel!("Lattice Dimension")
xlabel!("Wavevector")
```

# Make MStats Plot (Single Variable)

```julia
# load the data

@load "test_M_list.jld2" M_list;

fname = "test.json"
dat = Dict()
open(fname, "r") do f
    global dat
    dat=JSON.parse(f);  # parse and transform data
end
```

```julia
idx_list, d = make_idx(dat["vars"], dat)

# set Mstats options
options1 = Mstats_options()
options1["jld2"] = true;
options1["M_list"] = M_list;
options1["min_cutoff"] = 0.02;

M_stats = getMstats(dat, options1);
M_list_scaled = Array{Any}(undef, d)
scale_list = Array{Any}(undef, d);

for i in idx_list
    mstats_h = M_stats[string(i)]
    m_list_h = M_list[i]
    if isempty(mstats_h["vals"])
        max = maximum(broadcast(abs, m_list_h))[1]
        scale_list[i] = max;
    else
        max = maximum(mstats_h["vals"])
        scale_list[i] = max;
    end
    M_list_scaled[i] = 0.5*M_list[i]./max
end

options1 = Mstats_options()
options1["jld2"] = true;
options1["M_list"] = M_list_scaled;
options1["min_cutoff"] = 0.02;

M_stats = getMstats(dat, options1);
```

```julia
# set the variable to plot vs
dat["vars"] = ("α", "τ", "f")
var_key = "τ"
var_units = "";
subplot_idxs = [2; 10; 15]

# set the slice
dummy_idx = [3,30]
println(dat["α"][dummy_idx[1]])

# set options and make the plot
options = make_Mstats_options(dat, var_key, var_units, dummy_idx, subplot_idxs)
options["scaling"] = true
options["scale_list"] = 0.5 ./scale_list
p = make_Mstats_plot(dat, M_stats, subplot_idxs, 1, dummy_idx, options)

display(p)
```

```julia
function make_options_2d(dat, var1_key, var2_key)
        
    idx_list, d = make_idx(dat["vars"], dat)
    
    f = Dict()
    f["var1_key"] = var1_key
    f["var2_key"] = var2_key
    f["ylabel"] = var1_key
    f["xlabel"] = var2_key
    f["var1"] = dat[var1_key]
    f["var2"] = dat[var2_key]
    f["var1idx"] = findall(y -> y == var1_key, dat["vars"])[1]
    f["var2idx"] = findall(y -> y == var2_key, dat["vars"])[1]
    f["scale_list"] = ones(d)

    return f
end


function make_Mstats_plot_2d(dat, M_stats, slice_idx, options)
    
    var1 = options["var1"]
    var2 = options["var2"]
    
    M_max_surf = zeros(length(var1),length(var2))
    symm_surf = zeros(size(M_max_surf))
    num_peaks_surf = zeros(size(M_max_surf))
    ratio_surf = zeros(size(M_max_surf))
    
    idx_list, d = make_idx(dat["vars"], dat)
    
    var1idx = options["var1idx"]
    var2idx = options["var2idx"]
    
    for i in idx_list
        
        # make sure you're in the right slice of the data
        slice = [];
        for idx = 1:length(i)
            if idx != var1idx
                if idx != var2idx
                    push!(slice, i[idx])
                end
            end
        end

        if slice != slice_idx
            continue
        end
        
        # slice indexing
        i_slice = ();
        for idx = 1:length(i)
            
            if idx == var1idx
                i_slice = (i_slice..., i[idx])
            elseif idx == var2idx
                i_slice = (i_slice..., i[idx])
            end
            
        end
        
        if var1idx > var2idx
            i_slice = (i_slice[2], i_slice[1])
        end
        
        var1_h = var1[i[var1idx]]
        var2_h = var2[i[var2idx]]
        stats_h = M_stats[string(i)]
        min_idx = stats_h["min_idx"]
        max_idx = stats_h["max_idx"]
        sigmas = stats_h["sigmas"]
        vals = stats_h["vals"]
        symms = stats_h["symms"]
        M = stats_h["M"]
        t = stats_h["t"]
        τ = stats_h["τ"]
        nτ = stats_h["nτ"]

        if length(max_idx) == 0
            continue
        end
        peak_idx = findall(max_idx .== maximum(max_idx))

        M_max_surf[i_slice...] = maximum(vals*options["scale_list"][i]/0.5)
        symm_surf[i_slice...] = 1 - symms[peak_idx][1]
        num_peaks_surf[i_slice...] = length(vals)
        ratio_surf[i_slice...] = vals[peak_idx][1]/sigmas[peak_idx][1]
        
    end

    p1 = heatmap(var2, var1, M_max_surf, title="Max. Peak Height")
    p2 = heatmap(var2, var1, symm_surf, title="Asymmetry")
    p3 = heatmap(var2, var1, num_peaks_surf, title="Num. Peaks")
    p4 = heatmap(var2, var1, ratio_surf, title="Max. Peak Height/σ")
    
    ylabel!(p1, options["ylabel"])
    ylabel!(p3, options["ylabel"])
    
    xlabel!(p3, options["xlabel"])
    xlabel!(p4, options["xlabel"])
    
    l = @layout grid(2,2)
    
    plt = plot(p1, p2, p3, p4, layout = l, size = (800, 600))
    
    return plt

end
```

```julia
var1_key = "τ"
var2_key = "α"
slice_idx = [30]
println(dat["τ"][slice_idx[1]])

options = make_options_2d(dat, var1_key, var2_key)
options["scale_list"] = scale_list;
p = make_Mstats_plot_2d(dat, M_stats, slice_idx, options)
display(p)
```
