# Testing Different Shufflings

```julia
@load "003_M_list.jld2" M_list;
@load "003_parameters.jld2" params;
```

```julia
σ = ();
μ = ();
k = [x[2] for x in params["k"]];

for i = 1:20
    
    S = M_list[i,:];
    rS = Array{Any}(undef, 942, 400);
    for i = 1:400
        rS[:,i] = S[i];
    end
    rS = broadcast(abs, rS)

    meanS = sum(rS, dims = 2)./400;
    δ = Array{Any}(undef, 942, 400)

    for i = 1:400
        δ[:,i] = (rS[:,i] .- meanS).^2
    end

    σS = Array{Any}(undef, 942);

    for i = 1:942
        σS[i] = std(rS[i,:])/meanS[i]
    end

    σ = (σ..., σS);
    μ = (μ..., meanS)
    
end
```

```julia
i = 20;
println(k[i])
S = M_list[i,:];
rS = Array{Any}(undef, 942, 400);
for i = 1:400
    rS[:,i] = S[i];
end
rS = broadcast(abs, rS)
t = LinRange(0,300,942)
plot(t, μ[i])
# plot!(t, σ[i]*maximum(μ[i])/maximum(σ[i]))
plot!(t, rS[:, sample(1:400, 5, replace = false)])
```

```julia
t = LinRange(0, 300, 942)

p = plot()
for i = 20:-3:1
    S = M_list[i,:];
    rS = Array{Any}(undef, 942, 400);
    for i = 1:400
        rS[:,i] = S[i];
    end
    rS = broadcast(abs, rS)
    σS = σ[i]
    μS = μ[i]
    samp = rS[:, sample(1:400, 5, replace = false)];
    plot!(t, k[i] .+ 0 .* collect(t), σS, leg = false, yscale=:log10, lw=2, label=false, camera=(25,65), ticks = false)
    # plot!(t, k[i] .+ 0 .* collect(t), σS, leg = false, yscale=:log10, lw=2, label=false, camera=(25,65), ticks = false)
    plot!(t, k[i] .+ 0 .* collect(t), μS[:,1], leg = false, yscale=:log10, lw=2, label=false, camera=(25,65), ticks = false)
end

display(p)
```
